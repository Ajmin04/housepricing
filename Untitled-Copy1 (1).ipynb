# IQR FUNCTION

```python
Q1 =df['price_per_sqft'].quantile(0.25)
Q3 =df['price_per_sqft'].quantile(0.75)
IQR =Q3-Q1
Q1
Q3
IQR
```

```python
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
lower,upper
```
